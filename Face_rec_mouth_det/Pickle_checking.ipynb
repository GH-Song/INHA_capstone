```python
import pandas as pd

metadata = pd.read_pickle('output/recorded_mouth_data.pickle')
print("[INFO] Read DataFrame from pickle...")
print(metadata)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
p1 = fig.add_subplot(2,2,1)
p2 = fig.add_subplot(2,2,2)
p3 = fig.add_subplot(2,2,3)
p4 = fig.add_subplot(2,2,4)

data = metadata[metadata["record_time"] != 0]
speaking = data[data["Label"] == "speaking"]
x = speaking["record_time"]
y = speaking["InnerMAR"]
p1.plot(x,y, 'b-', label = "speaking")
p1.set_title("Innermouth_aspect_ratio")


y = y/y.mean()
p3.plot(x,y, 'b-', label = "speaking")
p3.set_title("Innermouth_aspect_ratio/mean")

nonspeaking = data[data["Label"] == "nonspeaking"]
x = nonspeaking["record_time"]
y = nonspeaking["InnerMAR"]
p1.plot(x,y, 'r-', label = "nonspeaking")


y= y/y.mean()
p3.plot(x,y, 'r-', label = "nonspeaking")

p1.legend(loc='best')
p3.legend(loc='best')
```

```python
x = speaking["record_time"]
y = speaking["0-6"]
p2.plot(x,y, 'b-', label = "speaking")

x = nonspeaking["record_time"]
y = nonspeaking["0-6"]
p2.plot(x,y,'r-', label = "nonspeaking")

x = speaking["record_time"]
y = speaking["0-6"]
y = y / y.mean()
p4.plot(x,y, 'b-', label = "speaking")

x = nonspeaking["record_time"]
y = nonspeaking["0-6"]
y = y/ y.mean()
p4.plot(x,y,'r-', label = "nonspeaking")

p2.legend(loc='best')
p4.legend(loc='best')
```

```python
p2.set_title("\nOuttermouth_width")
p4.set_title("Outtermouth_width/mean")
fig.suptitle("GH's recorded data")
fig.tight_layout()
```
